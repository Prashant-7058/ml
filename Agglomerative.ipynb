```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
X = df.iloc[ : , 3: ]

from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X = sc.fit_transform(X)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

lm = linkage(X, method='ward')
dendrogram(lm)
plt.title("Dendrogram")
plt.xlabel("Cutomers")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
Ag = AgglomerativeClustering(n_clusters=5, linkage='ward')
y_pred = Ag.fit_predict(X)
```

```python
plt.scatter(X[y_pred == 0,0], X[y_pred == 0,1], color='red')
plt.scatter(X[y_pred == 1,0], X[y_pred == 1,1], color='blue')
plt.scatter(X[y_pred == 2,0], X[y_pred == 2,1], color='yellow')
plt.scatter(X[y_pred == 3,0], X[y_pred == 3,1], color='violet')
plt.scatter(X[y_pred == 4,0], X[y_pred == 4,1], color='green')
plt.title("Clusters", fontsize=15)
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.show()
```
