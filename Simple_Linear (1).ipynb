```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('salary_ML.csv')
df
```

```python
X = df.iloc[ : , 0:1].values
y = df.iloc[ : , -1].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
X_pred = lr.predict(X_train)
```

```python
from sklearn.metrics import r2_score
R2Score = r2_score(y_test, y_pred)
print("R2Score is :", R2Score)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, X_pred, color='red')
plt.title("LinearRegression", fontsize=15)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='red')
plt.title("LinearRegression", fontsize=15)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error: ", mae)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("placement.csv")
```

```python
X = df.iloc[ : , 0:-1]
y = df.iloc[ : , 1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
X_pred = lr.predict(X_train)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, y_pred)
print("r2_score is :", r2_score)
```

```python
plt.scatter(X_train, y_train)
plt.plot( X_train, X_pred, color='red', linewidth=2)
plt.title("cgpa vs placement(training)")
plt.xlabel("cgpa")
plt.ylabel("placement")
plt.show()
```

```python
plt.scatter(X_test, y_test)
plt.plot( X_test, y_pred, color='red', linewidth=2)
plt.title("cgpa vs placement(testing)")
plt.xlabel("cgpa")
plt.ylabel("placement")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(mse)
```
