```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('50_Startups.csv')
```

```python
X = df.iloc[ : , :-1]
y = df.iloc[ : , -1]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
cs = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(cs.fit_transform(X))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3)
```

```python
from sklearn.linear_model import LinearRegression
rg = LinearRegression()
rg.fit(X_train, y_train)
```

```python
y_pred = rg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test, y_pred)
print("R2Score:", r2)
```

```python
import matplotlib.pyplot as plt
A = [1,2,3,4,5,6,7,8]
B = [3,4,5,6,7,9,9,8]
```

```python
plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test),max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('50_Startups.csv')
df
```

```python
X = df.iloc[ : , 0:4]
y = df.iloc[ : , -1]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
cl = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [3])], remainder ='passthrough')
X = np.array(cl.fit_transform(X))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, y_pred)
print("r2_score is :", r2_score)
```

```python
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)
```
